<h2> Edge Preserving Filter </h2>

where,<br>
src Input 8-bit 3-channel image <br>
dst Output 8-bit 3-channel image. <br> 
flags Edge preserving filters. <br>
It takes value RECURS_FILTER ( Recursive Filtering ) = 1 and NORMCONV_FILTER(<br>
Normalized Convolution ) = 2. Using RECURS_FILTER option is about 3.5x faster than NORMCONV_FILTER. The <br>NORMCONV_FILTER produces sharpening of the edges and it good for applications like stylizations. When sharpening is not desirable, and speed is important RECURS_FILTER should be used.  
<br>
sigma_s Range between 0 to 200 (See details below)
<br>sigma_r Range between 0 to 1 (See details below) 
<br>
sigma_s : controls the size of the neighborhood, <br>
sigma_r :(for sigma_range) controls the how dissimilar colors within the neighborhood will be averaged. <br>
A larger sigma_r results in large regions of constant color.<br>


```python
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 16,9

image = cv2.imread("./images/Manali-min.jpg")
print(image.shape)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

```python
dst = cv2.edgePreservingFilter(image, sigma_s=5, sigma_r=1)
plt.imshow(cv2.cvtColor(dst, cv2.COLOR_BGR2RGB))
```

<h2> Detail Enhancing Filter ( detailEnhance ) </h2>

```python
img = cv2.detailEnhance(dst, sigma_s=10, sigma_r=0.15)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_gray, img_color = cv2.pencilSketch(img, sigma_s=50, sigma_r=0.03, shade_factor=0.07)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
```

```python
styleImage = cv2.stylization(image, sigma_s=60, sigma_r=0.09)
plt.imshow(cv2.cvtColor(styleImage, cv2.COLOR_BGR2RGB))
```
